```python
import numpy as np 
from matplotlib import pyplot as plt 
student=[]
marks=[]


count=int(input("Enter Total records"))
for i in range(count):
    name=input("Enter  Student {} name".format(i+1))
    student.append(name)
    mrks=float(input("Enter Student {} Marks".format(i+1)))
    marks.append(mrks)
    
studentArray=np.array(student)
marksArray=np.array(marks)

plt.bar(studentArray,marksArray)
plt.show()
    

```

```python
# sub plots
import numpy as np 
from matplotlib import pyplot as plt

#plot 1
year=np.array([2014,2015,2016,2017,2018])
sales=np.array([2000, 3000, 4000, 3500, 6000])
purchase=np.array([1200, 1300, 1400, 1350, 1600])
plt.subplot(2,2,1)
plt.plot(year,sales)
plt.title("Sales Income")
plt.xlabel('Year')
plt.ylabel('Sales')

#plot2
plt.subplot(2,2,2)
plt.plot(year,purchase)
plt.title("Purchase Details")
plt.xlabel('Year')
plt.ylabel('purchase')
plt.show()

#plot3
profit=sales-purchase
plt.subplot(2,2,3)
plt.scatter(year,sales,s=profit/10)
plt.title("Profit Details")
plt.xlabel('Year')
plt.ylabel('sales')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

city=np.array(['Jalandhar','Ludhiana','Amritsar','Phagwara'])
population=np.array([10,15,20,5])
explodes=[0.1,0,0,0]
def func(pct):
    return "{:1.2f}%".format(pct)


plt.pie(population,labels=city,autopct=lambda pct:func(pct),explode=explodes,shadow=True)
plt.legend(title="cities")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
product=np.array(['A','B','C','D'])
sales=np.asarray([1000,2000,1200,2000])
profit=np.asarray([120,200,150,100])
plt.subplot(2,1,1)
colors=['red','b','g']
plt.bar(product,sales,width=0.2,color=colors)
plt.xticks(product)
plt.yticks(sales)
plt.title("Product Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.plot(product,sales,marker='o',linestyle='dashed')
plt.show()

plt.subplot(2,1,2)
plt.bar(product,profit,width=0.2,color=colors)
plt.xticks(product)
plt.yticks(profit)
plt.title("Product Profit")
plt.xlabel("Product")
plt.ylabel("Profit")
plt.show()
```
